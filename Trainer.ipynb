```python
import openai
import pandas as pd
from os import environ

openai.api_key = TELEGRAM_BOT_TOKEN = environ['OPENAI_API_KEY']
```

```python
df = pd.read_csv('data/processed/merged.csv')
# Format the data for fine-tuning
formatted_data = []
for i, row in df.iterrows():
    prompt = row['Code']  # Input text
    target = row['English']  # Expected output text
    formatted_data.append({
        'prompt': prompt,
        'target': target
    })
```

```python
formatted_data
```

```python
model = "text-davinci-003"
engine = "munin"
batch_size = 4
epochs = 3

response = openai.FineTune.create(
    prompt_samples=formatted_data,
    model=model,
    engine=engine,
    batch_size=batch_size,
)

print("Fine-tuning job ID: ", response.id)
```
